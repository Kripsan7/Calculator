```python
import math

def calculator():
    history = []
   
    print("--Welcome to Calculator--")
    while True:
        print("Select an option: ")
        print("1. Simple Calculations")
        print("2. Scientific Functions")
        print("3. Menu Actions")
        print("4. Unit Converter")

        pick = input("Enter your need: ")

        if pick == '1':
            print("Select an operation: ")
            print("1. Addition")
            print("2. Subtraction")
            print("3. Multiplication")
            print("4. Division")
            print("5. Power")
            print("6. Modulus")

            choice_1 = input("Enter a choice: ")

            if choice_1 not in ('1', '2', '3', '4', '5', '6'):
                print("Invalid choice. Try again.")
                continue

            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))

            except ValueError:
                print("Please enter valid numbers.")
                continue

            if choice_1 == '1':
                result = num1 + num2
                operation = f"{num1} + {num2} = {result}"

            elif choice_1 == '2':
                result = num1 - num2
                operation = f"{num1} - {num2} = {result}"

            elif choice_1 == '3':
                result = num1 * num2
                operation = f"{num1} * {num2} = {result}"
            elif choice_1 == '4':
                if num2 == 0:
                    print(" Cannot divide by zero!")
                    continue
                result = num1 / num2
                operation = f"{num1} / {num2} = {result}"
            elif choice_1 == '5':
                result = num1 ** num2
                operation = f"{num1} ^ {num2} = {result}"
            elif choice_1 == '6':
                result = num1 % num2
                operation = f"{num1} % {num2} = {result}"

            print(f" Result: {operation}")
            history.append(operation)
            continue

        elif pick == '2':
            print("Select an operation: ")
            print("1. Sine")
            print("2. Cosine")
            print("3. Tangent")
            print("4. Square root")
            print("5. Log(base 10)")
            print("6. Natural log")
            print("7. Factorial")

            choice_2 = input("Enter a choice: ")
            num = float(input("Enter a number: "))
            
            if choice_2 == '1':
                result = math.sin(math.radians(num)) # radians converts degree to radians
                operation = f"sin({num}) = {result}"

            elif choice_2 == '2':
                result = math.cos(math.radians(num))
                operation = f"cos({num}) = {result}"

            elif choice_2 == '3':
                result = math.tan(math.radians(num))
                operation = f"tan({num}) = {result}"

            elif choice_2 == '4':
                if num < 0:
                    print("Cannot take square root of a negative number.")
                    continue
                result = math.sqrt(num)
                operation = f"Square root of {num} = {result}"

            elif choice_2 == '5':
                if num <= 0:
                    print("Log base 10 is only defined for positive numbers.")
                    continue
                result = math.log10(num)
                operation = f"log10({num}) = {result}"

            elif choice_2 == '6':
                if num <= 0:
                    print("Natural log is only defined for positive numbers.")
                    continue
                result = math.log(num)
                operation = f"ln({num}) = {result}"

            elif choice_2 == '7':
                if num < 0 or not num.is_integer():
                    print("Factorial is only defined for non-negative integers.")
                    continue
                result = math.factorial(int(num))
                operation= f"{int(num)}! = {result}"

            print(f"Result: {operation}")
            history.append(operation)
            continue

        elif pick == '3':
            print("Select a command: ")
            print("1. View History")
            print("2. Clear History")
            print("3. Save History to File")
            print("X. Exit")

            choice_3 = input("Enter a choice: ").lower()
            if choice_3 == 'x':
                print("Exiting Calculator... Sayonara!")
                break

            elif choice_3 == '1':
                if not history:
                    print("No history yet.")
                else:
                    print(" Calculation History:")
                    for i, record in enumerate(history, 1):
                        print(f"{i}. {record}")
                continue


            elif choice_3 == '2':
                confirm = input("Are you sure you want to clear history? (y/n): ").lower()
                if(confirm == 'y'):
                    history.clear()
                    print("History cleared.")
                else:
                    print("Clear cancelled.")

                continue

            elif choice_3 == '3':
                if not history:
                    print("No history to save.")

                else:
                    filename = input("Enter the file name: ")+ ".txt"
                    file = open(filename, "w")

                    for item in history:
                        file.write(item +"\n")
                        file.close()
                        print(f"History written successfully to {filename}")

                continue

            continue

        elif pick == '4':
            print("Choose a Temperature Conversion: ")
            print("1. Temperature")
            print("2. Length")
            print("3. Weight")

            choice_4 = input("Enter a choice: ")
            if choice_4 == '1':
                print("Your options are:")
                print("1. Celsius to Fahrenheit")
                print("2. Fahrenheit to Celsius")
                print("3. Celsius to Kelvin")
                print("4. Kelvin to Celsius")

                temp_choice = float(input("Enter a choice: "))
                try:
                    temp = input("Enter the temperature: ")
                    
                    if temp_choice == '1':
                        result = (temp * 9/5) + 32
                        operation = f"{temp}°C = {result}°F"

                    elif temp_choice == '2':
                        result = (temp - 32) * 5/9
                        operation = f"{temp}°F = {result}°C"

                    elif temp_choice == '3':
                        result = temp + 273.15
                        operation = f"{temp}°C = {result}K"

                    elif temp_choice == '4':
                        result = temp - 273.15
                        operation = f"{temp}K = {result}°C"

                    else:
                        print("Invalid choice.")
                        continue

                except ValueError:
                    print("Invalid input.")
                    continue

            elif choice_4 == '2':
                print("Choose a Length Conversion:")
                print("1. Meters to Feet")
                print("2. Feet to Meters")
                print("3. Centimeters to Inches")
                print("4. Inches to Centimeters")
                length_choice = input("Enter a choice: ")
                
                try: 
                    length = float(input("Enter length: "))
                    if length_choice == '1':
                        result = length * 3.28084
                        operation = f"{length} meters = {result} feet"

                    elif length_choice == '2':
                        result = length / 3.28084
                        operation = f"{length} feet = {result} meters"

                    elif length_choice == '3':
                        result = length / 2.54
                        operation = f"{length} cm = {result} inches"

                    elif length_choice == '4':
                        result = length * 2.54
                        operation = f"{length} inches = {result} cm"

                    else:
                        print("Invalid choice.")
                        continue

                except ValuError: 
                    print("Invalid input.")
                    continue

            elif choice_4 == '3':
                print("Choose a Weight Conversion:")
                print("1. Kilograms to Pounds")
                print("2. Pounds to Kilograms")
                weight_choice = input ("Enter a choice: ")
                try:
                    weight = float(input("Enter weight: "))

                    if weight_choice == '1':
                        result = weight * 2.20462
                        operation = f"{weight} kg = {result} 1lbs"

                    elif weight_choice == '2':
                        result = weight / 2.20462
                        operation = f"{weight} lbs = {result} kg"

                    else:
                        print("Invalid choice")
                        continue

                except ValueError:
                    print("Invalid input.")
                    continue

            else:
                print("Invalid category.")
                continue

            print(f"Result: {operation}")
            history.append(operation)
            continue
                    
            

# Run the calculator
calculator()
     
        

   
     
```
